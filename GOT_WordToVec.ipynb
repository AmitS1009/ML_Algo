```python
import zipfile
import os

# Define paths
zip_file_path = '/content/archive.zip'  # Path to the zip file
unzip_folder_path = '/content/unzipped'  # Path where files will be unzipped

# Unzipping the file
def unzip_file(zip_path, extract_to):
    # Create the target directory if it doesn't exist
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)
        print(f"Created folder: {extract_to}")

    # Open and extract the zip file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"Unzipped files to: {extract_to}")

# Call the function
unzip_file(zip_file_path, unzip_folder_path)
```

```python
import os
import nltk
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

# Download punkt and punkt_tab tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')

story = []

# Path of the folder containing the files
folder_path = '/content/unzipped'

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # Open and read the file
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        corpus = f.read()

        # Sentence tokenization
        raw_sent = sent_tokenize(corpus)

        # Word tokenization and preprocessing
        for sent in raw_sent:
            story.append(simple_preprocess(sent))

# Displaying the first few tokenized words from the story
print(story[:50])  # Displaying first 50 words for verification
```

```python
story[:100]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
