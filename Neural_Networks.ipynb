```python
def load_dataset(dataset):
  if dataset == 'moons':
    X,y = make_moons(n_samples=10000, noise=0.15, random_state=1)

  elif dataset == 'circles':
      X,y = make_circles(n_samples=10000, noise=0.1, random_state=1, factor = 0.5)

  elif dataset == 'classification':
        X, y = make_classification(n_samples=10000, n_features=2, random_state=2, n_redundant=0)

  return X,y
```

```python
dataset = 'classification'
X, y = load_dataset(dataset)
X.shape, y.shape
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

### **Model Building : **

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(2,)))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model with loss, optimizer, and metrics
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display the model summary
model.summary()
```

```python
model.fit(X,y, epochs=100, batch_size=64, validation_split=0.1)
```
